```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of values from -10 to 10, which will serve as the x-axis values
x = np.linspace(-10, 10, 400)

# Calculate the logistic function (sigmoid) values for each x
y = 1 / (1 + np.exp(-x))

# Creating the plot
plt.figure(figsize=(6, 3))
plt.plot(x, y, label='f(x) = 1 / (1 + e^(-x))')
plt.title('Graph of the Logistic/Sigmond Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axhline(1, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-0.1, 1.1)
plt.xlim(-10, 10)
plt.legend()
plt.show()
```
